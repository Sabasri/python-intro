```python
fig, ax = plt.subplots()
ax.plot(x,y, label='linear')
```

```python
plt.close()
```

```python
x= np.arange(0,10,0.2)
y_sin =np.sin(x)
y_cos=np.cos(x)
```

```python
plt.plot(x,y_sin,label='sine',linewidth=4.0)
plt.plot(x,y_cos,label='cos')
plt.plot(x,y_cos, label='cos_2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plotting")
plt.legend()
plt.show()
plt.close()
```

multiple plots

```python
x=np.arange(0,10,0.2)
```

```python
y_sin =np.sin(x)
y_cos=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine Plot')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cos Plot')
```
